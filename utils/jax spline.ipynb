```python
import jax
import numpy as np
import jax.numpy as jnp
import jax_cosmo as jc
```

```python
jc.scipy.interpolate.InterpolatedUnivariateSpline
```

```python
x = np.arange(0, 2*np.pi+np.pi/4, 2*np.pi/16)
rng = np.random.default_rng()
y =  np.sin(x) + 0.1*rng.standard_normal(size=len(x))
```

```python
import matplotlib.pyplot as plt
xnew = np.arange(0, 9/4, 1/50) * np.pi
plt.plot(xnew, np.sin(xnew), '-.', label='sin(x)')
plt.plot(x, y, 'o')

tck = jc.scipy.interpolate.InterpolatedUnivariateSpline( x, y, k=3)
plt.plot( xnew, tck(xnew), 'r' )
```
